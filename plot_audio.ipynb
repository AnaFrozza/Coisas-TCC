# Espectro Chroma do audio

```python
ipt = 're.wav'
y, sr = librosa.load(ipt)
```

```python
ipd.Audio(y, rate=sr)
```

```python
N = 1024
f_max = 17000

#Computar o espectro de potência do sinal de áudio
stft = np.abs(librosa.stft(y, n_fft=N, window='hann'))**2

mel_stft = librosa.feature.melspectrogram(S=stft, sr=sr, n_mels=128, fmax=f_max)

chroma_cqt = librosa.feature.chroma_cqt(y=y, sr=sr)

# plt.figure(1)
# #plt.subplot(3,1,1)
# librosa.display.specshow(librosa.power_to_db(mel_stft, ref=np.max), y_axis='mel', fmax=f_max, x_axis='time')
# plt.colorbar(format='%+2.0f dB')
# plt.title('Mel Spectrogram')
# plt.tight_layout()
# plt.show()

# plt.figure(2)
# #plt.subplot(3,2,1)
# librosa.display.specshow(librosa.power_to_db(stft, ref=np.max), y_axis='linear', fmax=f_max, x_axis='time')
# plt.colorbar(format='%+2.0f dB')
# plt.title('STFT')
# plt.tight_layout()
# plt.show()

# plt.figure(3)
# plt.subplot(3,1,3)

# Chromagrama por passo da classe
librosa.display.specshow(chroma_cqt, y_axis='chroma', x_axis='time')
# Espectrograma por Hz
# librosa.display.specshow(chroma_cqt, y_axis='cqt_hz')
# Espectrograma por Nota
# librosa.display.specshow(chroma_cqt, y_axis='cqt_note')


#Teste STFT
# D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
# Amplitudo linear dos DB da STFT
# librosa.display.specshow(D, y_axis='linear')
# Amplitudo em escala logaritimica dos DB da STFT
# librosa.display.specshow(D, y_axis='log')


plt.title('chroma_cqt')
plt.colorbar(format='%2f dB')
plt.tight_layout()
plt.show()
```

# Espectro Tempograma do audio

```python
ipt = 're.wav'
x, sr = librosa.load(ipt)
```

```python
ipd.Audio(x, rate=sr)
```

```python
#Temporal
# plt.subplot(4, 2, 8)
Tgram = librosa.feature.tempogram(y=x, sr=sr)
librosa.display.specshow(Tgram, x_axis='time', y_axis='tempo')
plt.colorbar()
plt.title('Tempogram')
plt.tight_layout()
```
